```python
# from above graph we can clearly say that it is very imbalanced data as perspective of "Class" Column
# so for that we need to balanced it using smote technique wich used for minority oversampling 
```

```python
X = df.drop("Class", axis=1)
Y = df['Class']
```

```python
# Split the dataset into training data and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
# we need to balanced the dataset
```

```python
# for that we use the SMOTE technique.
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)

x_train, y_train = smote.fit_resample(X_train,Y_train)
x_test, y_test = smote.fit_resample(X_test,Y_test)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = LogisticRegression(max_iter=100)
model.fit(x_train, y_train)
```

```python
y_pred_test=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
matrix=confusion_matrix(y_test, y_pred_test)
sns.heatmap(matrix,annot=True)
plt.show()
print(matrix)
```
